```python
# =========== sciaganie i laczenie informacji o notowaniach funduszy ===========
wstep = "https://stooq.pl/q/d/l/?s="
zakonczenie_fundusz = ".n&i=d&c=1" 

linki_z_url = []
for tiker, nazwa_funduszu in tikery_dict.items():
    url = wstep + tiker + zakonczenie_fundusz
    try:
        df = pd.read_csv(url, delimiter=";", parse_dates=["Data"])

        df["link"] = url
        df["Nazwa_Funduszu"] = nazwa_funduszu

        linki_z_url.append(df)
        liczba_wierszy = len(df)
        
        opoznienie = random.uniform(1, 4)
        print(f"Pobrałem << {nazwa_funduszu} >> (liczba wierszy: {liczba_wierszy}), czekam {opoznienie:.2f} sekund...")
        time.sleep(opoznienie)

    except Exception as e:
        print(f"Nie udało się pobrać danych z linku {url} dla tickera {tiker} - błąd: {e}")

if linki_z_url:
    polaczone_dane = pd.concat(linki_z_url, ignore_index=True)
```

```python
# =========== ODPYTYWANIE GEMINI (PRZEKAZYWANIE MU KART FUNDUSZU JAKO TEKST I GENEROWANIE ODPOWIEDZI NA PYTANIA)  ===========
import pymupdf4llm
from google import genai

WSTEP_SCIEZKI = r"abc"
FOLDER_WYNIKOW_GEMINI = r"abc"

API_KEY_GEMINI = "abc"
MODEL_GEMINI = "gemini-2.5-flash"
# wczesniej (przed 22.06.25): gemini-2.5-flash-preview-04-17

# =========== Inicjalizacja klienta Gemini ===========
try:
    client = genai.Client(api_key=API_KEY_GEMINI)
    print("INFO: Klient Gemini został pomyślnie zainicjowany.")
except Exception as e:
    print(f"KRYTYCZNY BŁĄD: Nie można zainicjować klienta Gemini: {e}")
    client = None

# =========== Główna logika przetwarzania ===========
if client:
    os.makedirs(FOLDER_WYNIKOW_GEMINI, exist_ok=True)
    
    for tiker in os.listdir(WSTEP_SCIEZKI):
        sciezka_funduszu_folder = os.path.join(WSTEP_SCIEZKI, tiker)

        if os.path.isdir(sciezka_funduszu_folder):
            print(f"\n--- Przetwarzanie folderu tickera: {tiker} ---")

            sciezka_folderu_wynikow_tikera = os.path.join(FOLDER_WYNIKOW_GEMINI, tiker)
            os.makedirs(sciezka_folderu_wynikow_tikera, exist_ok=True)

            for data_element in lista_dat:
                
                nazwa_pliku_wynikowego_txt = data_element + ".txt"
                sciezka_do_pliku_wynikowego_txt = os.path.join(sciezka_folderu_wynikow_tikera, nazwa_pliku_wynikowego_txt)

                if os.path.exists(sciezka_do_pliku_wynikowego_txt):
                    # print(f"INFO: Plik wynikowy {sciezka_do_pliku_wynikowego_txt} już istnieje. Pomijam przetwarzanie.")
                    continue

                sciezka_funduszu_plik = os.path.join(sciezka_funduszu_folder, data_element) + ".pdf"
                print(f"Próba przetworzenia pliku PDF: {sciezka_funduszu_plik}")

                if not os.path.exists(sciezka_funduszu_plik):
                    print(f"OSTRZEŻENIE: Plik PDF {sciezka_funduszu_plik} nie istnieje. Pomijam.")
                    continue

                try:
                    md_text = pymupdf4llm.to_markdown(sciezka_funduszu_plik,
                                                    ignore_images=True,
                                                    ignore_graphics=True)

                    if not md_text.strip():
                        print(f"OSTRZEŻENIE: Brak tekstu w pliku PDF {sciezka_funduszu_plik} po konwersji. Pomijam.")
                        continue

                    zapytanie_do_gemini = f"""
Przesyłam Ci przykładowy tekst z PDF'a:
--- POCZĄTEK TEKSTU Z PDF ---
{md_text}
--- KONIEC TEKSTU Z PDF ---
Jest to przykładowa karta funduszu i chciałbym, abyś odpowiedział na pytania:
1. Jaka jest wartość aktywów netto (prosba o podanie wartosci w mln; dodaj proszę mln PLN na końcu, np. 1000 mln PLN, 515 mln PLN)?
2. Jaki jest ustawiony benchmark (bez zbędnego info, czyli np. '100% WIG20', '75% WIG20 + 25% mWIG40')?
3. Jaka jest struktura aktywów (instrumenty finansowe, ile procent, bez zbędnego info, w stylu: 'Akcje 97,96%, Inne aktywa 2,04% @-@' itp.)?
4. Jaka jest struktura geograficzna (kraje, ile procent, bez zbędnego info)?
5. Jakie są największe pozycje w portfelu (nazwy spółek, obligacji, ile procent, bez zbędnego info, w stylu: 'Spółka X 10%, Spółka Y 20%', itp)?
6. Jaka jest alokacja sektorowa (nazwy branż, ile procent, bez zbędnego info)?
7. Jakie waluty są w funduszu (nazwy, jak 'EURO', 'PLN' itp, ile procent, bez zbędnego info)
--- DODATKOWE WAŻNE INFORMACJE CO DO ODPOWIEDZI ---
Chciałbym abyś przesłał tylko odpowiedzi wg numeru pytania, bez innych informacji jak "Oto odpowiedzi:". 
Jeżeli nie jesteś w stanie znaleźć info, zwróć 'Brak'.
Bardzo proszę abyś zachował spójność, tzn. jak przesyłasz odpowiedz, zaczynaj od '1. tekst', potem zacznij od nowej linijki '2. tekst' i tak przez każdą odpowiedź.
Nie dodawaj spacji po %, tzn. tak nie: 90,15 % , tylko tak: 90,15%.
Jeżeli piszesz coś łącznie, tzn. jest benchmark który w jednym ciągu ma '+', to prośba o oddzielenie ich w ten sposób: ' + ' czyli spacja plus spacja.
Części dziesiętne rozdzielaj przecinkiem, np. 99,15 bez kropek.
Prześlij odpowiedź na każde pytanie w jednej linii, tzn. nie rób tak: 3. Anglia 5%, nowa linijka i Czechy 10%, tylko w jednej linii: Anglia 5%, Czechy 10% itp.
Jak skończysz odpowiadać na dane pytanie, zakończ to tym '@-@', czyli np. 1. 150 mln PLN @-@ nowa linijka 2. Wig20 @-@ nowa linijka itp.
"""

                    odpowiedz_od_gemini = client.models.generate_content(
                        model=MODEL_GEMINI,
                        contents=[zapytanie_do_gemini]
                    )

                    print(f"\nODPOWIEDŹ GEMINI dla: {tiker} / {data_element}.pdf")
                    
                    tekst_do_zapisu = ""
                    if hasattr(odpowiedz_od_gemini, 'text'):
                        tekst_do_zapisu = odpowiedz_od_gemini.text
                        print(tekst_do_zapisu)
                    elif hasattr(odpowiedz_od_gemini, 'parts') and odpowiedz_od_gemini.parts:
                        tekst_do_zapisu = odpowiedz_od_gemini.parts[0].text
                        print(tekst_do_zapisu)
                    else:
                        tekst_do_zapisu = f"Surowa odpowiedź: {odpowiedz_od_gemini}"
                        print(tekst_do_zapisu)
                    
                    try:
                        with open(sciezka_do_pliku_wynikowego_txt, 'w', encoding='utf-8') as plik_txt:
                            plik_txt.write(tekst_do_zapisu)
                        print(f"Odpowiedź zapisana do: {sciezka_do_pliku_wynikowego_txt}")
                    except Exception as e_zapis:
                        print(f"BŁĄD podczas zapisywania pliku {sciezka_do_pliku_wynikowego_txt}: {e_zapis}")

                except Exception as e:
                    print(f"BŁĄD podczas przetwarzania pliku PDF {sciezka_funduszu_plik}: {e}")

                # Ten sleep wykona się, jeśli nie było 'continue' wcześniej w tej iteracji
                print("Czekam 4 sekundy...")
                time.sleep(4)
        else:
            print(f"INFO: Pomijam element (nie jest folderem): {sciezka_funduszu_folder}")

print("\n--- Zakończono przetwarzanie wszystkich folderów i dat ---")
```

```python
# =========== Ogarnianie danych z kart funduszu do excela ===========
lista_an = []
lista_bench = []
lista_sa = []
lista_sg = []
lista_npp = []
lista_as = []
lista_w = []

for ticker_id, ticker_dokumenty in tikery_dict_extra_info.items():
    ticker_dokumenty = tikery_dict_extra_info[ticker_id]["ticker_dokumenty"]
    sciezka_do_zwrotu_gemini = fr"abc\{ticker_dokumenty}"

    for data in lista_dat:
        sciezka_do_zwrotu_gemini_pliki_txt = fr"abc\{ticker_dokumenty}\{data}.txt"
        try:
            df_all = pd.read_csv(sciezka_do_zwrotu_gemini_pliki_txt, sep="\t", header=None)

            # podchodze do tego indywidualnie, by miec otwarta furtke na przyszlosc
            # ====== AKTYWA NETTO ======
            df_an = df_all.iloc[0]
            df_an = df_an.str.replace("1.", "").str.replace("mln PLN @-@", "").str.replace(" ", "").str.replace(",", ".").str.strip().astype(float)
            df_an = df_an.to_frame(name="aktywa_netto")
            df_an["tiker"] = ticker_id
            df_an["data"] = data
            lista_an.append(df_an)

            # ====== BENCHMARK ======
            df_bench = df_all.iloc[1]
            df_bench = df_bench.str.replace("2.", "").str.replace("@-@", "").str.strip()
            df_bench = df_bench.to_frame(name="benchmark")
            df_bench["tiker"] = ticker_id
            df_bench["data"] = data
            lista_bench.append(df_bench)

            # ====== STRUKTURA AKTYWOW ======
            df_sa = df_all.iloc[2]
            df_sa = df_sa.str.replace("3.", "").str.replace("@-@", "").str.strip()
            df_sa = df_sa.str.split(", ")
            df_sa = df_sa.to_frame(name="struktura_aktywow")
            df_sa["tiker"] = ticker_id
            df_sa["data"] = data
            lista_sa.append(df_sa)

            # ====== STRUKTURA GEOGRAFICZNA ======
            df_sg = df_all.iloc[3]
            df_sg = df_sg.str.replace("4.", "").str.replace("@-@", "").str.strip()
            df_sg = df_sg.str.split(", ")
            df_sg = df_sg.to_frame(name="struktura_geograficzna")
            df_sg["tiker"] = ticker_id
            df_sg["data"] = data
            lista_sg.append(df_sg)

            # ====== NAJW POZYCJE W PORTFELU ======
            df_npp = df_all.iloc[4]
            df_npp = df_npp.str.replace("5.", "").str.replace("@-@", "").str.strip()
            df_npp = df_npp.str.split(", ")
            df_npp = df_npp.to_frame(name="najw_poz_w_portfelu")
            df_npp["tiker"] = ticker_id
            df_npp["data"] = data
            lista_npp.append(df_npp)

            # ====== ALOKACJA SEKTOROWA ======
            df_as = df_all.iloc[5]
            df_as = df_as.str.replace("6.", "").str.replace("@-@", "").str.strip()
            df_as = df_as.str.split(", ")
            df_as = df_as.to_frame(name="alokacja_sektorowa")
            df_as["tiker"] = ticker_id
            df_as["data"] = data
            lista_as.append(df_as)

            # ====== WALUTY ======
            df_w = df_all.iloc[6]
            df_w = df_w.str.replace("7.", "").str.replace("@-@", "").str.strip()
            df_w = df_w.str.split(", ")
            df_w = df_w.to_frame(name="waluty")
            df_w["tiker"] = ticker_id
            df_w["data"] = data
            lista_w.append(df_w)

        except Exception as e:
            print(f"Wystąpił inny błąd: {e} - {ticker_dokumenty} - {data}")

aktywa_netto_df = pd.concat(lista_an, ignore_index=True)
aktywa_netto_df.to_excel(r"abc\aktywa.xlsx")

benchmark_df = pd.concat(lista_bench, ignore_index=True)
benchmark_df.to_excel(r"abc\benchmark.xlsx")

struktura_aktywow_df = pd.concat(lista_sa, ignore_index=True)
struktura_aktywow_df.to_excel(r"abc\struktura_aktywow.xlsx")

struktura_geograficzna_df = pd.concat(lista_sg, ignore_index=True)
struktura_geograficzna_df.to_excel(r"abc\struktura_geograficzna.xlsx")

najw_poz_w_portfelu_df = pd.concat(lista_npp, ignore_index=True)
najw_poz_w_portfelu_df.to_excel(r"abc\najw_poz_w_porfelu.xlsx")

alokacja_sektorowa_df = pd.concat(lista_as, ignore_index=True)
alokacja_sektorowa_df.to_excel(r"abc\alokacja_sektorowa.xlsx")

waluty_df = pd.concat(lista_w, ignore_index=True)
waluty_df.to_excel(r"abc\waluty.xlsx")
```
